Машин сургалтын сан, фрэмворкууд
Регресс


```python
from sklearn.linear_model import LinearRegression

# Сургалтын өгөгдөл (X=оролт, y=гаралт)
X = [[1], [2], [3], [4]]

y = [2, 4, 6, 8]

# Загвар үүсгэх ба сургах
model = LinearRegression()
model.fit(X, y)
```

```python
# Таамаглах (predict)
p = model.predict([[5]])
print('p : ',p)
# [10.]
```

Ангилал

Жимс, хүнсний нэр + жин, бүтэц → Тэр нь алим уу, жүрж үү гэдгийг таамаглана.

```python
from sklearn.tree import DecisionTreeClassifier

# Өгөгдөл: [жин (gram), гадаргуу: 0-гөлгөр, 1-барзгар]
X = [
    [150, 0],  # Алим
    [170, 0],  # Алим
    [140, 1],  # Жүрж
    [130, 1]   # Жүрж
]

# Шошго (label): 0-Алим, 1-Жүрж
y = [0, 0, 1, 1]

# Загвар үүсгэх
model = DecisionTreeClassifier()
model.fit(X, y)
```

```python
# Шинэ жимс: 160гр, гөлгөр – энэ нь юу вэ?
prediction = model.predict([[160, 0]])
print("Таамаг:", "Алим" if prediction[0] == 0 else "Жүрж")
```

Кластер

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans

# 2 хэмжээст өгөгдөл: (X, Y) координаттай цэгүүд
data = np.array([
    [1, 2], [1.5, 1.8], [5, 8],
    [8, 8], [1, 0.6], [9, 11]
])

# 2 бүлэг болгон ангилна
kmeans = KMeans(n_clusters=2)
kmeans.fit(data)

# Төвүүд болон кластерт хуваарилалт
centroids = kmeans.cluster_centers_
labels = kmeans.labels_
```

```python
# Цэгүүдийг кластерийн дугаараар өнгөөр ялгаж зурна
colors = ["r", "g"]  # улаан ба ногоон
for i in range(len(data)):
    plt.scatter(data[i][0], data[i][1], color=colors[labels[i]])

# Кластерын төвүүдийг харуулна
plt.scatter(centroids[:, 0], centroids[:, 1], marker='x', s=200, linewidths=3, color='black')
plt.title("K-Means Кластэрийн үр дүн")
plt.show()
```
