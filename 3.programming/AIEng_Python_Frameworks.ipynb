```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# 1. Өгөгдөл үүсгэх (NumPy ашиглан)
sales = np.array([100, 150, 200, 130, 180])
```

```python
# 2. DataFrame болгож боловсруулах (pandas ашиглан)
df = pd.DataFrame({
    'Day': ['Mon', 'Tue', 'Wed', 'Thu', 'Fri'],
    'Sales': sales
})
```

```python
# 3. Харагдацыг гаргах (matplotlib ашиглан)
plt.plot(df['Day'], df['Sales'], marker='o')
plt.title("Өдөр бүрийн борлуулалт")
plt.xlabel("Өдөр")
plt.ylabel("Борлуулалт")
plt.grid(True)
plt.show()
```
